```python
import pandas as pd
from textblob import TextBlob
import sqlite3

# 1. Data load 
df = pd.read_csv('data.csv')

# 2. (Reviews + Ratings + Departments)
# 'Department Name' aur 'Class Name'  original file 
cols_to_keep = ['Review Text', 'Rating', 'Department Name', 'Class Name']
df = df[cols_to_keep].dropna()

# 3. Sentiment Analysis Function
def get_sentiment(text):
    score = TextBlob(str(text)).sentiment.polarity
    if score > 0: return 'Positive'
    elif score < 0: return 'Negative'
    else: return 'Neutral'

# 4. Sentiment apply
df['Sentiment'] = df['Review Text'].apply(get_sentiment)

# 5. New CSV file save (Yahi file Power BI mein jayegi)
df.to_csv('cleaned_sentiment_data.csv', index=False)

# 6. SQL Database update 
conn = sqlite3.connect('ecommerce_project.db')
df.to_sql('reviews_table', conn, if_exists='replace', index=False)

print("columns:", df.columns.tolist())
```
